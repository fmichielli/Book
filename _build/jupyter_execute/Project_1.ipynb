## Part 2: Data Analysis without Pandas

Read in data as a list of dictionaries and look at first values:
* Split each row into columns with the delimeter ",", since the datat is read in as a CSV
* This will give individual values for each category (or column) per row

```python
data = []
with open ("DHS_Daily_Report_20241028.csv", 'r') as DHS_data:
    keys = DHS_data.readline().rstrip('\n').split(",")
    data1 = DHS_data.readlines()

    for row in data1:
        row = row.rstrip('\n')
        rowlist = row.split(",")
        dict = {}
        for i in range(len(keys)):
            dict[keys[i]] = rowlist[i]
        data.append(dict)
data[1]
```

Select a column and create a list of its values, then preview the list:

```python
adult_pop = []
for entry in data: 
    adults = float(entry['Total Adults in Shelter'])
    adult_pop.append(adults)

adult_pop[:10]
```

__Find the mean:__
* Sum the values and divide by a count of the values

```python
sum = 0
count = 0
for n in adult_pop:
    sum += n
    count += 1
mean = round(sum/count,2)
print (f'The mean total adults in shelter is {mean}')
```

Find the median:
* sort the list
* find the middle value or the two middle values of the list (and take the midpoint)

```python
adult_pop.sort()
if len(adult_pop)%2 == 0:
    dist = len(adult_pop)/2
    median = (adult_pop[int(dist)] + adult_pop[int(dist-1)])/2
else: 
    median = adult_pop[int((len(adult_pop)-1)/2)]

print (f'The median total adults in shelter is {median}')
```

__Find the mode:__
* Get set of unique values
* Count the number of time each value occurs in the list
* Report the value with the highest number of occurrences
* If there are multiple modes this will report all 

```python
unique = set(adult_pop)
largest = 0
mode_set = {}
for n in unique:
    count = 0
    for i in adult_pop:
        if i == n:
            count += 1
    if count > largest:
        largest = count
        mode_set[n] = count
    if count == largest: 
        mode_set[n] = count
        
mode = []
for p in mode_set:
    if mode_set[p] == largest:
        mode.append(p)

print (f'The mode total adults in shelter is {mode}')
```

## Part 3: Data Visualization

Import data:

```python
import pandas as pd

DHS_data = pd.read_csv('DHS_Daily_Report_20241028.csv')
DHS_data.head() 
```

Create a "Year" column:

```python
DHS_data['Year'] = pd.DatetimeIndex(DHS_data['Date of Census']).year
DHS_data.info()
```

```python
DHS_data.head()
```

Create a set of years covered by the data:

```python
years = set(DHS_data["Year"])
years
```

Create new dataframe that is grouped by year to get average daily adult population in shelter by year:

```python
DHS_year = DHS_data.groupby("Year").mean(True).reset_index()
```

```python
DHS_year
```

Create a dictionary: key is year, value is total adults in shelter. 

Then sort by year (which creates a list of tuples).

```python
adults_per_year = {}
for i in years:
    adults_per_year[i] = DHS_year.loc[DHS_year["Year"] == i, 
    "Total Adults in Shelter"].iloc[0]

adults_per_year = sorted(adults_per_year.items())

adults_per_year
```

Convert mean yearly adult shelter population into thousands for each year 
and print as sparkline:

```python
print ("Average Daily Total Adults in Shelter in Thousands, by Year: ")
for key in adults_per_year:
    adjusted = int(round(key[1]/1000,0))
    stars = "*" * adjusted
    print (f'{key[0]}: {stars}')
```
